```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Случайная величина X имеет нормальное распределение, M[X] = 10 и D[X] = 4.


```python
M = 10
D = 4
E = np.sqrt(D)
```

1. Построить функцию и плотность распределения вероятностей для нормированной нормально распределенной случайной величины

```python
array_dots = np.linspace(-4, 4, 100)
pdf_norm = stats.norm.pdf(array_dots)
cdf_norm = stats.norm.cdf(array_dots)

plt.plot(array_dots, pdf_norm, 'b-', linewidth=2, label='Плотность вероятности')
plt.plot(array_dots, cdf_norm, 'r-', linewidth=2, label='Функция распределения')
plt.title('Нормированное нормальное распределение', fontsize=14, fontweight='bold')
plt.xlabel('z', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.xlim(-4, 4)
```

Построить функцию и плотность распределения вероятностей для X.

```python
array_x_dots = np.linspace(M - 4*E, M + 4*E, 100)
pdf_x_norm = stats.norm.pdf(array_x_dots, M, E)
cdf_x_norm = stats.norm.cdf(array_x_dots, M, E)

plt.plot(array_x_dots, pdf_x_norm, 'b-', linewidth=2, label='Плотность вероятности')
plt.plot(array_x_dots, cdf_x_norm, 'r-', linewidth=2, label='Функция распределения')
plt.title('Исходное распределение', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
```

Вычислить вероятности P(X < 11), P(X > 5), P(4 < X < 12)

```python
p_less_11 = stats.norm.cdf(11, M, E)
p_more_5 = 1 - stats.norm.cdf(5, M, E)
p_between_4_12 = stats.norm.cdf(12, M, E) - stats.norm.cdf(4, M, E)

print(f'P(X < 11) = {p_less_11}\n'
      f'P(X > 5) = {p_more_5}\n'
      f'P(4 < X < 12) = {p_between_4_12}')

plt.plot(array_x_dots, pdf_x_norm, 'b-', linewidth=2)

mask_less_11 = array_x_dots < 11
plt.fill_between(array_x_dots[mask_less_11], pdf_x_norm[mask_less_11], 
                 alpha=0.4, color='green', label=f'P(X < 11) = {p_less_11:.3f}')

mask_more_5 = array_x_dots > 5
plt.fill_between(array_x_dots[mask_more_5], pdf_x_norm[mask_more_5], 
                 alpha=0.3, color='orange', label=f'P(X > 5) = {p_more_5:.3f}')

mask_between_4_12 = (array_x_dots > 4) & (array_x_dots < 12)
plt.fill_between(array_x_dots[mask_between_4_12], pdf_x_norm[mask_between_4_12], 
                 alpha=0.5, color='red', label=f'P(4 < X < 12) = {p_between_4_12:.3f}')

plt.xlabel('x', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11) 
```

Проверить выполнение правила «3-сигм» для X, отобразить это правило на графике плотности вероятностей

```python
up = M + 3 * E
down = M - 3 * E
p_3_sigma = stats.norm.cdf(up, M, E) - stats.norm.cdf(down, M, E)

plt.plot(array_x_dots, pdf_x_norm, 'b-', linewidth=2, label='Плотность вероятности')
mask_3_sigma = (array_x_dots >= down) & (array_x_dots <= up)
plt.fill_between(array_x_dots[mask_3_sigma], pdf_x_norm[mask_3_sigma], alpha=0.6, color='lightgreen', label='Область 3-сигм')
plt.axvline(down, color='red', linestyle='--', alpha=0.7)
plt.axvline(up, color='red', linestyle='--', alpha=0.7)
plt.title('Правило "3-сигм"', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
```

```python
M_Y = 2
lambd = 1 / M_Y
```

Построить для нее функцию и плотность распределения вероятностей

```python
array_y_dots = np.linspace(0, 5 * M_Y, 100)
pdf_y = stats.expon.pdf(array_y_dots, scale=M_Y)
cdf_y = stats.expon.cdf(array_y_dots, scale=M_Y)

plt.plot(array_y_dots, pdf_y, 'b-', linewidth=2, label='Плотность вероятности')
plt.plot(array_y_dots, cdf_y, 'r-', linewidth=2, label='Функция распределения')

plt.title('Экспоненциальное распределение', fontsize=14, fontweight='bold')
plt.xlabel('y', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
```

Вычислить вероятности P(X < 3), P(X > 2), P(1 < X < 2.5)
